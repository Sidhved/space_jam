```python
pop_list = []

for id in df['ID']:
    if id <= 30:
        pop_list.append('player'+str(id))
    else:
        pop_list.append('tune_squad'+str(id%30))

pop_list
```

```python
df['player'] = pop_list
df.head()
```

```python
column_list = list(df.iloc[:, :-1])
column_list.insert(1, 'player')
df= df[column_list]
df.head()
```

```python
df[['GP', 'MPG']] = df[['GP', 'MPG']].fillna(value=df[['GP', 'MPG']].mean())
df.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

X = df.dropna(how= 'any').iloc[:, 5:-1].to_numpy()
y = df.dropna(how='any').iloc[:, -1]

lin_reg = LinearRegression()
scores = cross_val_score(lin_reg, X, y, cv=10, scoring='r2')
entries, bin_edges, patches = plt.hist(scores, bins = 10);

print('Mean r2 score: {:.4f}'.format(scores.mean()))
for i in range(len(entries)):
    if entries[i]>0:
        print('{:.0f}% of r2 scores are between {:.4f} and {:.4f}'.format(entries[i]*100/len(entries), bin_edges[i], bin_edges[i+1]))
```

```python
lin_reg.fit(X, y)
```

```python
mask = df.isnull().any(axis=1)
mask
```

```python
df.loc[mask].iloc[:, 5:-1]
```

```python
df.loc[mask, 'PER'] = lin_reg.predict(df.loc[mask].iloc[:, 5:-1])
df.isna().sum()
```
